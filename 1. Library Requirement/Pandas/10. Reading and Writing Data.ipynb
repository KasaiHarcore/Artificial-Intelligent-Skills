```python
import pandas as pd  
import numpy as np 
```

```python
df2=pd.read_table("Data/data2.txt", sep=",")
df2
```

```python
df=pd.read_table("Data/data2.txt", sep=",")
df
```

```python
df=pd.read_table("Data/data2.txt",
                 sep=",",
                 header=None)
df
```

```python
df=pd.read_table("Data/data2.txt",
                 sep=",",
                 header=None, 
                 names=["name","score","sex"])
df
```

```python
df=pd.read_table("Data/data2.txt",
                 sep=",",header=None, 
                 names=["name","score","sex"],
                 index_col="name")
df
```

```python
df2=pd.read_table("Data/data3.txt",
                  sep=",")
df2
```

```python
df2=pd.read_table("Data/data3.txt",
                  sep=",",
                  index_col=["lesson","name"])
df2
```

```python
df3=pd.read_table("Data/data4.txt",sep=",")
df3
```

```python
df3=pd.read_table("Data/data4.txt",
                  sep=",", 
                  skiprows=[0,2])
df3
```

```python
df3=pd.read_table("Data/data4.txt",
                  sep=",", 
                  skiprows=[0,2],
                  usecols=[0,1])
df3
```

```python
df3=pd.read_table("Data/data4.txt",
                  sep=",", 
                  skiprows=[0,2],
                  usecols=[0,1],
                  nrows=3)
df3
```
