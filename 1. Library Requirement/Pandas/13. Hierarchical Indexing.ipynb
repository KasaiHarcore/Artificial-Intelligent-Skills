# Hierarchical Indexing in Pandas

```python
import numpy as np
import pandas as pd
```

```python
data=pd.Series(np.random.randn(8),
               index=[["a","a","a","b",
                       "b","b","c","c"],
                      [1,2,3,1,2,3,1,2]])
data
```

```python
data.index
```

```python
data["a"]
```

```python
data["b":"c"]
```

```python
data.loc[["a","c"]]
```

```python
data.loc[:,1]
```

```python
data.unstack()
```

```python
data.unstack().stack()
```

```python
df=pd.DataFrame(
    np.arange(12).reshape(4,3),
    index=[["a","a","b","b"],
           [1,2,1,2]],
    columns=[["num","num","ver"],
             ["math","stat","geo"]])
df
```

```python
df.index.names=["class","exam"]
df.columns.names=["field","lesson"]
df
```

```python
df["num"]
```

```python
df.swaplevel("class","exam")
```

```python
df.sort_index(level=1)
```

## Summary Statistics by Level

```python
df.sum(level="exam")
```

```python
df.sum(level="field",axis=1)
```

## Indexing with a DataFrame’s columns

```python
data2=data.set_index(["a","b"])
data2
```
