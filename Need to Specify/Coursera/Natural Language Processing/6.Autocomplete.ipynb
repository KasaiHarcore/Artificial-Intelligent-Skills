### Xc sut Trigram

 tnh xc sut, ta c cng thc sau:
- $P(w_3|w_1^2) = \frac{count(w_1^2*w_3)}{count(w_1^2)}$
- $Count(w_1^2*w_3) = Count(w_1*w_2*w_3) = Count(w_1^3)$

Xc sut N-gram:
- $P(w_N | w_1^{N-1}) = \frac{count(w_1^{N-1}w_N)}{count(w_1^{N-1})}$
- $Count(w_1^{N-1}w_N) = Count(w_1^{N})$

### Xc sut Trnh t (Sequence Probability)

Cc kt qu ngn ca chng v cui cng l cch tnh gn ng xc sut N-gram. Khi lm nh vy, ta c gng c tnh xc sut ca mt cu. Vi cu "Ti tn l t", ta c xc sut ca cu ny l:
- $P(B | A) = \frac{P(A, B)}{P(A)} \rightarrow P(A, B) = P(A) * P(B | A)$
- $P(A, B, C, D) = P(A) * P(B | A) * P(C | A, B) * P(D | A, B, C)$

T , ta c:
- $P(Ti, tn, l, t) = P(Ti) * P(tn | Ti) * P(l | Ti, tn) * P(t | Ti, tn, l)$

Mt trong nhng vn  ln nht vi cch tnh ny l corpus him khi cha cc cm t ging ht nh trn. Do , ta rt d c kh nng nhn c xc sut l 0. Gi nh v Markov ch ra rng t cui cng mi quan trng. Do vy:
- Bigram: $P(w_N | w_1^{N-1}) = P(w_N | w_{N-1})$
- N-gram: $P(w_N | w_1^{n-1}) = P(w_N | w_{n - N + 1}^{n-1})$

T  ta c m hnh:
- $P(w_i^N) \approx \Pi_{i=1}^{N} P(w_i | w_{i - 1})$
- $P(w_i^N) \approx P(w_1) * P(w_2|w_1) * ... * P(w_N|w_{n - 1})$
