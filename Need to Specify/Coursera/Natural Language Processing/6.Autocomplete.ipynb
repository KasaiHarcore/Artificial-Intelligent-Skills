# N-Grams Overview

N-gram là cơ bản và cung cấp cho bạn nền tảng cho phép bạn hiểu các mô hình phức tạp hơn trong chuyên ngành. Những mô hình này cho phép bạn tính toán xác suất của một số từ nhất định xảy ra theo một trình tự cụ thể. Bằng cách đó, bạn có thể xây dựng một công cụ tự động sửa lỗi hoặc thậm chí là một công cụ gợi ý tìm kiếm.

Các ứng dụng bao gồm:



### Xác suất và N-grams

Trước khi bắt đầu tính toán xác suất của các chuỗi nhất định, trước tiên chúng ta cần xác định mô hình ngôn ngữ N-gram là gì:



Với những định nghĩa đó, ta có thể gán nhãn cho một câu như sau



Với:
- $w_1^{i} = w_1 * w_2 * w_3 * ... * w_i$ 
- $w_1^{3} = w_1 * w_2 * w_3$
- $w_{m - 2}^{i} = w_{m - 2} * w_{m - 1} * w_i$

VD: Ta có mỗi chuỗi sau: "Xin chào, tôi tên là Đạt"
- Kích cỡ 1-gram: "Xin", "chào", "tôi", "tên", "là", "Đạt"; m = 6
- P("tôi") = 1/6
- P("tên") = 1/6

Tổng hợp, ta có xác suất của unigram là: $P(w) = \frac{count(w)}{m}$

### Xác suất Trigram

Để tính xác suất, ta có công thức sau:
- $P(w_3|w_1^2) = \frac{count(w_1^2*w_3)}{count(w_1^2)}$
- $Count(w_1^2*w_3) = Count(w_1*w_2*w_3) = Count(w_1^3)$

Xác suất N-gram:
- $P(w_N | w_1^{N-1}) = \frac{count(w_1^{N-1}w_N)}{count(w_1^{N-1})}$
- $Count(w_1^{N-1}w_N) = Count(w_1^{N})$

### Xác suất Trình tự (Sequence Probability)

Các kết quả ngắn của chúng và cuối cùng là cách tính gần đúng xác suất N-gram. Khi làm như vậy, ta cố gắng ước tính xác suất của một câu. Với câu "Tôi tên là Đạt", ta có xác suất của câu này là:
- $P(B | A) = \frac{P(A, B)}{P(A)} \rightarrow P(A, B) = P(A) * P(B | A)$
- $P(A, B, C, D) = P(A) * P(B | A) * P(C | A, B) * P(D | A, B, C)$

Từ đó, ta có:
- $P(Tôi, tên, là, Đạt) = P(Tôi) * P(tên | Tôi) * P(là | Tôi, tên) * P(Đạt | Tôi, tên, là)$

Một trong những vấn đề lớn nhất với cách tính này là corpus hiếm khi chứa các cụm từ giống hệt như trên. Do đó, ta rất dễ có khả năng nhận được xác suất là 0. Giả định về Markov chỉ ra rằng từ cuối cùng mới quan trọng. Do vậy:
- Bigram: $P(w_N | w_1^{N-1}) = P(w_N | w_{N-1})$
- N-gram: $P(w_N | w_1^{n-1}) = P(w_N | w_{n - N + 1}^{n-1})$

Từ đó ta có mô hình:
- $P(w_i^N) \approx \Pi_{i=1}^{N} P(w_i | w_{i - 1})$
- $P(w_i^N) \approx P(w_1) * P(w_2|w_1) * ... * P(w_N|w_{n - 1})$
