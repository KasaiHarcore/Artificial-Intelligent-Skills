### Biến đổi word vector

Theo như bài trước, ta đã hiểu về cách tạo ra các vector cho từng từ và cách mapping giúp ta dịch ngôn ngữ sử dụng ma trận biến đổi (transformation matrix)



Trong tiếng Pháp, chat có nghĩa là cat trong tiếng Anh. Ta có thể dịch từ đó bằng cách nhân vector của từ chat với ma trận biến đổi và sau đó sử dụng cosin similarity để tìm từ gần nhất trong từ điển tiếng Anh.

X là vector của từ tiếng anh, Y là tiếng Pháp, R là ma trận biến đổi.

Cách học ma trận R:
- Khởi tạo ma trận R ngẫu nhiên
- For loop:
    + $Loss = \sum_{i=1}^{m} ||X^{(i)}R - Y^{(i)}||^2$
    + $g = \frac{\partial Loss}{\partial R}$
    + $R = R - \alpha g$

Mục tiêu là tối thiểu hóa hàm loss.

### Hash tables và Hash functions

Giả sử ta phải phân loại các hình sau vào các hộp chứa tương ứng. Ta có thể sử dụng hash function để giúp ta phân loại nhanh chóng.



Bạn có thể coi hash function như là một hàm lấy dữ liệu kích thước và mapping nó tới một giá trị cố định:



Với ví dụ trên, ta có 10 hộp và 5 dữ liệu là [10, 14, 97, 17 ,100] và phải phân loại chúng vào các hộp. 

Với 100, ta có thể sử dụng hash function là h(x) = x mod 10 = 0 => 100 sẽ vào hộp 0.

Với 14, ta có thể sử dụng hash function là h(x) = x mod 10 = 4 => 14 sẽ vào hộp 4.

Tương tự như vậy với các số còn lại|

### Locally sensitive hashing

Đây là một kỹ thuật giúp ta hash các dữ liệu giống nhau vào cùng một khối với xác suất cao.



Thay vì sử dụng phương pháp thông thường, ta có thể phân cụm các điểm giữ liệu bằng cách quết định xem chúng có nằm trên hay nằm dưới đường thẳng. Ta hãy xét tới dữ liệu nhiều chiều, ta sẽ sử dụng mặt phẳng thay vì đường thẳng:



Tại điểm P(1, 1), ta sẽ sử dụng dot product giữa các vector và điểm đó. Từ đó ta có thể thấy rằng điểm P(1, 1) sẽ vuông góc với mặt phẳng (nét đứt). Bất kì vector nào phía trên nét đứt khi nhân với (1, 1) sẽ là dương, còn lại sẽ là âm. Các vector nằm trên đường thẳng nhân với (1, 1) sẽ bằng 0.

Cách thể hiện hình chiếu:



Khi dot product giữa 2 vector, sau đó lấy độ lớn hoặc độ dài của vectơ đó, ta sẽ có được đường màu đen (được gắn nhãn là Phép chiếu). Dấu cho biết vectơ hình chiếu nằm ở phía nào của mặt phẳng.

### Đa mặt phẳng

Ta có thể áp dụng nhiều mặt phẳng để phân loại dữ liệu:



Cho một vài điểm biểu thị bằng **v**, ta chạy qua các phép chiếu $P_1, P_2, P_3$.

Với:
- $P_1v^T$ ta được một số dương, tức h ta sẽ set là 1.
- $P_2v^T$ cũng tương tự, ta sẽ set là 1.
- $P_3v^T$ ta sẽ set là 0.

Sau đó ta tính hash value:

$hash = 2^0 * 1 + 2^1 * 1 + 2^2 * 0 = 3$

Công thức tổng quát:

$hashvalue = \sum_{i=1}^{n} 2^{i-1} * 1_{P_i(v) > 0}$
