### Naive Bayes Model

Khi training, ta sẽ có 6 bước chính như sau:
- Đếm số lần xuất hiện của chữ
- Preprocess giống như LogReg
- Tính toán tần xuất xuất hiện của các từ

- Tính xác xuất số lần xuất hiện của chữ đó với mỗi class
- Tính ratio

- Tính log prior với $D_pos$ và $D_neg$ tương ứng với tổng số dữ liệu thuộc class positive và negative tương ứng.

### Ứng dụng của Naive Bayes

Có rất nhiều ứng dụng của Bayes ngây thơ bao gồm:
- Nhận dạng tác giả
- Lọc thư rác
- Truy xuất thông tin
- Định hướng từ

Phương pháp này thường được sử dụng như một đường cơ sở đơn giản. Nó cũng thực sự nhanh chóng.

### Assumptions của Naive Bayes

Naïve Bayes đưa ra giả định về tính độc lập và bị ảnh hưởng bởi tần suất từ trong kho ngữ liệu. Ví dụ: nếu bạn có những điều sau đây:

- It is sunny and hot in the Sahara desert
- It's alway cold and snowy in ___

Ở câu đầu tiên, bạn có thể thấy từ nắng và nóng có xu hướng phụ thuộc lẫn nhau và có mối tương quan ở một mức độ nhất định với từ "sa mạc". Naive Bayes đảm nhận sự độc lập xuyên suốt. Hơn nữa, nếu bạn điền vào câu bên phải, mô hình ngây thơ này sẽ gán trọng số bằng nhau cho các từ "xuân, hạ, thu, đông".
