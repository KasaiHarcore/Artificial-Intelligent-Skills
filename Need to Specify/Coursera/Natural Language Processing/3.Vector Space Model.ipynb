# Vector Space Models

Không gian vector là nền tảng trong nhiều ứng dụng của NLP. Nó rất quan trọng với các nhiệm vụ trích xuất thông tin, dịch thuật và chatbot. Ta có thể dùng các vector đại diện cho kí tự để biểu diễn mối quan hệ của nó.



Khi mô hình học các vector, ta thường vận dụng các từ ngữ có độ tương quan để rút ra ý nghĩa và thông tin.

### Word by Word và Word by Doc

Về thiết kế của WbW, ta sẽ bắt đầu bằng việc khám phá từng từ trong câu. Giả sử ta đang cố gắng tìm ra một vector cho một từ nhất định, ta sẽ tạo một ma trận mà trong đó hàng và cột tương ứng với các từ trong corpus.

Từ đó ta sẽ lặp lại liên tục với mỗi tài liệu khác nhau để theo dõi thông số trong ma trận đó.



Trong ví dụ trên, K ở đây đóng vai trò là khoảng cách (băng thông) quyết định sự liên kết của chúng

Còn WbD, tương tự với WbW, ta sẽ cũng sẽ cố tìm ra các từ và so sánh chúng với thông tin còn lại, nhưng đặc biệt hơn là ta so sánh từ đó với một tập tài liệu thay vì từng từ.

### Điều chỉnh từ ngữ trong không gian vector

Vectơ từ để thực sự trích xuất các mẫu và xác định các cấu trúc nhất định trong văn bản.

Ta có thể sử dụng từ vectơ cho Nga, Hoa Kỳ và DC để tính toán một vectơ rất giống với vectơ của Moscow. Sau đó, bạn có thể sử dụng độ tương tự cosine của vectơ với tất cả các vectơ từ khác mà bạn có và bạn có thể thấy rằng vectơ Moscow là gần nhất



### Visualize và PCA

PCA là một thuật toán unsupervised thường được dùng giảm kích thước dữ liệu, kết quả là nó có thể biểu diễn giữ liệu nhiều chiều dưới dạng 2 chiều. Sự kết hợp giữa variance với các feature.



Lưu ý rằng khi thực hiện PCA trên dữ liệu này, bạn sẽ thấy dầu khí ở gần nhau và thị trấn và thành phố cũng gần nhau. Để vẽ dữ liệu, bạn có thể sử dụng PCA để đi từ $d > 2$ kích thước để $d = 2$



### Tổng quát PCA

PCA thường được sử dụng để giảm kích thước dữ liệu của bạn. Theo trực giác, mô hình thu gọn dữ liệu trên các thành phần chính. Bạn có thể coi thành phần chính đầu tiên (trong tập dữ liệu 2D) là dòng có nhiều phương sai nhất. Sau đó bạn có thể thu gọn các điểm dữ liệu trên dòng đó. Do đó bạn đã chuyển từ 2D sang 1D. Bạn có thể khái quát trực giác này thành nhiều chiều.



Eigenvector: Các điểm không tương quan với dữ liệu

Eigenvalue: Thông tin được giữ lại

Các bước tính toán:
- Mean normalize dữ liệu
- Tính covariance
- Tính SVD
