Trong việc chỉnh sửa, ta có các thao tác thường dùng như là thêm, xóa, sửa, hoặc thay thế. Để xác định từ đúng, ta cần tìm chuỗi thao tác có độ dài nhỏ nhất để chuyển từ sai chính tả thành từ đúng. Đây chính là Minimum Edit Distance.

### Minimum Edit Distance Algorithm

Khi bắt đầu tính toán, ta sẽ tạo một ma trận có kích thước (m+1)x(n+1) với m và n là độ dài của từng từ. Ta sẽ điền vào ma trận này theo thứ tự từ trái qua phải, từ trên xuống dưới.



Từ đó ta sẽ có bảng giá trị sau khi điền vào theo việc thêm, xóa, hoặc thay thế.



Cách sử dụng này còn được gọi là Levenshtein Distance. Nếu xây dựng lại đường dẫn từ chuỗi này sang chuỗi khác, ta sẽ sử dụng backtrace. Ta sẽ đi từ phần tử cuối cùng của ma trận và đi ngược lại theo đường dẫn có giá trị nhỏ nhất.

Phương pháp tính toán thay vì brute force là môt một kỹ thuật khá phổ biến là Dynamic Programming. Đầu tiên ta giải quyết các bài toán con và sau đó sử dụng kết quả của các bài toán con đó để giải quyết bài toán lớn hơn.
