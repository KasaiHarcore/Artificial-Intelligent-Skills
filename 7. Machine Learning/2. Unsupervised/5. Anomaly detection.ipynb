In Unsupervised Learning algorithms, the main goal is to find the hidden patterns and structures in the data. Anomaly detection is a type of unsupervised learning where the main goal is to identify the data points that are significantly different from the rest of the data. Anomaly detection is also known as outlier detection. Anomaly detection is used in various applications such as fraud detection, network security, fault detection, and many more.

An anomaly can be broadly categorized into three categories:
- Point Anomaly: A tuple in a dataset is said to be a Point Anomaly if it is far off from the rest of the data.
- Contextual Anomaly: An observation is a Contextual Anomaly if it is an anomaly because of the context of the observation.
- Collective Anomaly: A set of data instances help in finding an anomaly.

First of all, we need to understand what is an anomaly. An anomaly is a data point that is significantly different from the rest of the data.

In practice, anomaly detection is often used to detect suspicious events, unexpected opportunities or bad data buried in time series data. A suspicious event might indicate a network breach, fraud, crime, disease or faulty equipment. An unexpected opportunity could involve finding a store, product or salesperson that's performing much better than others and should be investigated for insight into improving the business.

An anomaly could also be the result of faulty equipment, broken sensors or a disconnected network. In these instances, a data scientist might want to remove the anomalous data records from further analysis so as not to compromise the development of new algorithms.

### Supervised learning use cases

#### Retail
- Labeled data from previous years' sales can predict future sales goals, set benchmarks for sales employees, and analyze patterns for insights into products, marketing, and seasonality.

#### Weather Forecasting
- Supervised learning algorithms use historical data to predict weather patterns, analyzing recent barometric pressure, temperature, and wind speeds for more accurate forecasts.

**The most commonly used algorithms for this purpose are supervised Neural Networks, Support Vector Machine learning, K-Nearest Neighbors Classifier, etc.**

### Unsupervised learning use cases

#### Intrusion detection system
- Machine learning algorithms monitor network traffic for security violations or malicious activity. They can detect real-time attacks, protecting user information and system functions. These systems analyze time series data to identify spikes in traffic or unexpected patterns.

#### Manufacturing
- Unsupervised learning algorithms can be utilized for predictive maintenance of machinery, predicting potential failures or malfunctions from unlabeled data from equipment sensors, thereby reducing machine downtime.

**The most commonly used algorithms for this purpose are K-means clustering, DBSCAN, Isolation Forest, etc.**
