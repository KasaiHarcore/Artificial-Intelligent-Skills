# Basic

### Heading

```md
# Heading 1
## Heading 2
### Heading 3
#### Heading 4
##### Heading 5
###### Heading 6
```

### Text formatting

| Target        | Syntax                     | Results    |
| ------------- | -------------------------- | ---------- |
| Italic        | `*text*` or `_text_`       | *text*     |
| Bold          | `**text**` or `__text__`   | **text**   |
| Bold + Italic | `***text***`               | ***text*** |
| Strikethrough | `~~text~~`                 | ~~text~~   |

### Lists

No order lists

```
- Item 1
- Item 2
    - Sub Item
```

Order lists

```md
1. Item 1
1. Item 2
    1. Sub item
```

### Links

```md
[OpenAI](https://openai.com)
```

Or

```md
<https://openai.com>
```

### Images

```md
![Alt text](image_url)
```
