# Reading a txt file

We’ll also introduce several functions that provide useful information about the file.

In order to demonstrate how to read these files, we’re going to work with an example, which has the file path shown here. You can also take a quick look at the file’s contents. Our example file contains three lines of text from Wikipedia’s page.

```R
text <- readLines("Arthur.txt")

text

# [1] Success can mean a variety of different things. Success is, quite simply, the accomplishment of a predetermined goal. To some people it could mean making money, cultivate and develop certain basic qualities, to others it could mean keeping everyone happy, but to me, it means achieving the goals and objective I have set for myself for my life. Besides working on your goals that would lead a person towards success it is very important to push your limit every day, take charge of your life, and keep learning. This experience enables us to think smartly to solve a critical problem and achieve success. It is very important to take care of your mind which could be done by eliminating negative thoughts and negative people from your life. I think in order to call something successful, both the result and the process should be great. Without success, you, the group, your company, your goals, dreams and even entire civilizations cease to survive.
```

- There're some useful operations that we can perform on the file path. For example, we can get the name of the file using the basename function.

```R
basename("Arthur.txt")

# [1] "Arthur.txt"
```
- dirname returns the directory name of the file.

```R
dirname("Arthur.txt")

# [1] "."
```

- We can also get the absolute path of the file using the normalizePath function.

```R
normalizePath("Arthur.txt")

# [1] "/your/path/file/Arthur.txt"
```

- We can also check if a file exists using the file.exists function.

```R
file.exists("Arthur.txt")

# [1] TRUE
```

- To taken the file size, we can use the file.info function.

```R
file.size("Arthur.txt")

# [1] 1000
```

- There also another method call scan, which is used to read data into R. It’s a very flexible function, but it’s also a bit more complicated to use.

```R
scan("Arthur.txt", what = "character", sep = "\n")
```
