# Basic math with R

There is a lot of math that can be done with R. We will start with the basics.

## Basic math

R can be used as a calculator. The basic math operators are:

- `+` for addition
```R
1 + 1
```

- `-` for subtraction
```R
1 - 1
```

- `*` for multiplication
```R
2 * 3
```

- `/` for division
```R
4 / 2
```

- `^` for exponentiation
```R
2 ^ 8
```

- `%%` for modulo
```R
5 %% 2
```

- `%/%` for integer division
```R
5 %/% 2
```

## Variables

Variables are used to store values. You can assign a value to a variable using `<-` or `=`. For example:
```R
x <- 1
y = 1
```

You can then use the variable in place of the value:
```R
x + y
```

Variable can be overwritten like Python variables:
```R
x <- 2
x <- x + 1
x
```

To remove a variable, use `rm`:
```R
rm(x)
```

## Order of operations

The order of operations in R is the same as in Python. For example:
```R
total <- (193 + 2200) * 0.5
total
total_hr <- total / 60
total_hr
```
