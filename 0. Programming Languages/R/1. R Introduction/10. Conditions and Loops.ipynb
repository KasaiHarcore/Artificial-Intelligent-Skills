To do this, we’re going to have to rely on a conditional statement. In other words, we need to examine the dataset row by row, and check if the year value is greater than 2000 or not.
```R
if (condition) {
  # do something (if condition is TRUE)
} else {
    # do something else (if condition is FALSE)
}
```

## For loops
- Let's us take an example, where we want to print the numbers from 1 to 5. We can do this using a for loop. In R, we can use a for loop to iterate over a vector, a list, or a data frame. In this case, we want to iterate over a vector of numbers from 1 to 5. We can do this using the colon operator, which is a shortcut for creating a vector of numbers. We can then use the for loop to iterate over the vector and print each number.

```R
for (val in vector) {
  # do something
}
```

- Like python, the loop can be contain some conditional statement, even new vector, list, variable can be created inside the loop.

```R
years <- c(1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997,
           1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005,
           2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013,
           2014, 2015, 2016, 2017, 2018, 2019, 2020)

for (year in years) {
  if (year > 2000) {
    print(year)
  } else {
    print("This year is not greater than 2000")
  }
}
```

- Same as Python, we can use break and next to control the loop.

```R
i <- 1

while (i <= 5) {
  if (i == 3) {
    i <- i + 1
    next # Instead of (continue) in Python
  }
  print(i)
  if (i == 4) {
    break
  }
  i <- i + 1
}
```
