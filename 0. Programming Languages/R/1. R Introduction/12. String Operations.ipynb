# Basic String Operations

Strings are bits of text. They can be defined as anything between quotes:

```R
a_string <- "Hello"
another_string <- 'How are you?'
```

But to be easy, we will start with a txt file, to read it, we use the function `readLines()`:

```R
summary <- readLines("data/summary.txt") # Replace the path with your own
```

We can counting the number of characters in a string with the function `nchar()`:

```R
nchar(summary)
```

We can also use the function `substr()` to extract a substring from a string:

```R
substr(summary, 1, 10)
```

We can also use the function `paste()` to concatenate strings:

```R
paste("Hello", "World!")
```

We can also use the function `toupper()` to convert a string to uppercase:

```R
toupper(summary)
```

We can also use the function `tolower()` to convert a string to lowercase:

```R
tolower(summary)
```

We can also use the function `strsplit()` to split a string into substrings:

```R
strsplit(summary, " ")
```

We can also use the function `grep()` to search for a pattern in a string:

```R
grep("the", summary)
```

We can also use the function `gsub()` to replace a pattern in a string:

```R
gsub("the", "THE", summary)
```

We can also use the function `grepl()` to search for a pattern in a string and return a logical value:

```R
grepl("the", summary)
```

We can also use the function `sub()` to replace a pattern in a string:

```R
sub("the", "THE", summary)
```

We can also use the function `strtrim()` to trim a string:

```R
strtrim(summary, 10)
```

We can also use the function `strtoi()` to convert a string to an integer:

```R
strtoi("123")
```
