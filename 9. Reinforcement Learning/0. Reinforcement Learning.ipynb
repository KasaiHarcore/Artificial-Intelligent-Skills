# Introduction

To make it easier to visualize RL (Reinforcement Learning), I will give you a simple example as follows.

In 2100, where technology has advanced with robots, and you are the owner of a robot factory. And your biggest task is to train the robots to pick up the boxes and put them in the correct position.
- First, you help them how to place the boxes in the correct position. In each time the robot picks up the box and puts it down on the good spot or not, you will help them correct it. The robot just need to follow your instructions and remember it. This is called Supervised Learning.
- After a while, the robot has a new task: pick up the box and classify it into the correct category. You give it a bunch of boxes with different colors and shapes. And because your robot has learned how to place the boxes in the correct position, it use the previous knowledge to pick it up and try to figure out where the box should be placed. This is called Unsupervised Learning.
- Finally, you really proud of your robot and maybe give it a some encouragement. Now you take the robot to the warehouse, still with the same task, but this time the robot have to categorize more than 1000 box with thousand of features. You give it a reward when it does the right thing and a punishment or nothing when it does the wrong thing, this force the robot to find the best way to do the task. This is called Reinforcement Learning.

This is the most special part of RL that I want to talk about, when we have to teach AI how to operate in real environments.

Ok so I'm sure that you will have a bunch of questions like:
- Why Machine Learning also included Reinforcement Learning but we categorize it as a separate field?
- Why we have to learn Deep Learning before Reinforcement Learning?

First, Reinforcement Learning is a part of Machine Learning, but it is a separate field because it has a different way to learn and solve problems. In Supervised Learning and Unsupervised Learning, we have a dataset that contains the input and output data, and we use this data to train the model. But in Reinforcement Learning, we don't have the dataset, we have to interact with the environment to get the data.

Second, RL is a very complex fields, it requires a lot of knowledge about Machine Learning, Optimization, and Mathematics. And Deep Learning is a part of Machine Learning, it helps us to solve the complex problem with a lot of data. If we implement RL in Machine Learning, our brain will explode because we have to deal with a lot crazy things like advanced algorithms, theory, and heavy computation. So we need to learn Deep Learning first to understand how to deal with the complex problem and then we can learn RL.

### Reinforcement Learning

Before talking more about RL, we should understand that the word reinforce in RL means to strengthen, supplement, and complement. Therefore, we must also understand that this is a method of teaching models that can already function with the hope that it can become better, the purpose of RL up to now is not to replace learning methods such as SL or USL but is just a method born to supplement.

Reinforcement Learning is similar to other learning methods ending in '-ing', these are all continuous and possibly infinite processes. RL problems are built and developed to be able to map situations with actions with the goal of maximizing Rewards signals. In another understanding, RL is a closed-loop method with conditions:
- No specific instructions.
- Includes a series of actions and rewards.
- Ends with a specific condition.

There is a big difference between RL and today's popular learning methods, while SL or USL algorithms are based on a specific data set, RL is based on interactions between objects. This leads to RL being able to develop itself in many different situations without being completely dependent on the data provided.

One of the biggest challenges of RL is the trade-off between **exploration** and **exploitation**. In order to optimize the reward, the agent needs to make choices in whether to seek other directions or test old directions that are considered good. If we only choose to explore, the agent will not be able to find the best direction, and only choosing to exploit will limit the agent's capabilities. This is called the ***exploration-explotation dilemma***, this dilemma has been studied and researched up to now and coming up with a specific solution to balance exploration and exploitation is still a difficult problem.

**Note**: this dilemma does not arise in the supervised or unsupervised problem, at least in its purest form.

### Examples

Beside of my example above, there are many examples of RL that you can see in real life:
- AlphaGo: The first computer program to defeat a professional human Go player. It uses a combination of supervised learning from human expert games, and reinforcement learning from games it plays against itself.
- Dota 2: OpenAI Five, a team of five neural networks, defeated human professional players in the game Dota 2.
- An adaptive controller adjusts the temperature of a room by learning the temperature that maximizes the comfort of the occupants while minimizing energy usage. (Which not followed any specific instructions)
- An autonomous vehicle learns to drive by driving in the real world and observing the results of its actions. (Which already appear every where like Tesla, Waymo, etc.)


These examples share features that are so basic that they are often overlooked. All involve an agent interacting with an environment to achieve a goal. The agent chooses actions and the environment responds to these actions with rewards and new states, which are then observed by the agent. The agent then uses this information to choose its next action. This is the basic loop of RL.

At the same time, in all of these examples, the agent is not told which actions to take; thus the agent must monitor its environment and decide what to do to achieve its goal.

### Elements of Reinforcement Learning

Before talking about Agent and Environment, we will talk about the 4 main components in the RL system:
- ***Policy***: understood as the direction of the agent at a specific time, looking deeper, policy is the way we map the current state of the environment with a specific action. In psychology, this is called "Set of stimulus-response rules of associations". In fact, policy can be a basic equation, it can be a look-up table, in more complex cases it can be a series of complex calculation processes, ... In RL, policy is a subjective way for us to evaluate the behavior of the agent.
- ***Reward*** (Reward signal): This is considered the goal of RL, for each step, the environment will return to the agent a value called reward. The agent's goal will be to optimize the total reward in a series of actions (a long run). In a more general perspective, reward is a value that represents the good or bad state of the current environment for the agent. In biology, this is a good or bad experience for a problem. In reality, this is only a temporary experience and the reward will be considered to represent the immediate states.
- ***Value Function***: Unlike reward, value function is very suitable for long run. To be precise, the value of a state is the total amount of reward that the agent can achieve cumulatively in the future, starting from the current state. While the reward is determined immediately, the value is determined in the near or distant future. This leads to the possibility of a state with low reward but still high value. *Note: reward is a fixed value and value will have to be estimated and re-estimated based on the actions that the agent has taken.
- **Model of environment**: This is considered as the way we copy the state of the field, representing how the environment will behave. For example, when we have a pair of state and action, the model will predict which state will receive the reward next. Model is often used in planning, because we also need to consider whether performing an action will change the state in the future by considering the predicted state and reward. Methods that use models are called model-based methods, otherwise they are called model-free methods.
